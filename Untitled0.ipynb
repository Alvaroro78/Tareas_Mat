Ejercicio 1. Serie Fibonnaci.

```python
import matplotlib.pyplot as plt
serie_fibonaci = [0, 1]
def fibonacci(n):
  for i in range(2, n):
    serie_fibonaci.append(serie_fibonaci[i-1] + serie_fibonaci[i-2])

numeros = int(input("Digite la cantidad de números que desea generar de acuerdo a la serie de fibonnaci"))
if numeros <= 0:
  print("Digite un número mayor a 0")
elif numeros == 1:
  print(serie_fibonaci[0])
elif numeros == 2:
  print(serie_fibonaci)
else:
  fibonacci(numeros)

plt.figure(figsize=(8, 5))
plt.plot(range(numeros), serie_fibonaci, marker='o', linestyle='-', color='b', label="Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title(f"Serie de Fibonacci (Primeros {numeros} términos)")
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 2. Análisis de datos metereológicos.
DATOS: Temperaturas: [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

```python
import numpy as np
import matplotlib.pyplot as plt

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2,
                27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion_std = np.std(temperaturas)

print(f"Media de las temperaturas: {media:.2f}°C")
print(f"Mediana de las temperaturas: {mediana:.2f}°C")
print(f"Desviación estándar: {desviacion_std:.2f}°C")

meses = list(range(1, len(temperaturas) + 1))
plt.figure(figsize=(8, 5))
plt.plot(meses, temperaturas, marker='o', linestyle='-', color='r', label="Temp. media mensual")

plt.xlabel("Mes")
plt.ylabel("Temperatura (°C)")
plt.title("Temperatura Media Mensual")
plt.xticks(meses)
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 3. Simulación dado tradicional.

```python
import matplotlib.pyplot as plt
import random
lanzamientos = []
conteo = [0,0,0,0,0,0]
valores= [1,2,3,4,5,6]
for y in range(1000):
  lanzamiento = random.randint(1, 6)
  lanzamientos.append(lanzamiento)
for x in range(1000):
  conteo[lanzamientos[x]-2] = conteo[lanzamientos[x]-2] + 1

plt.figure(figsize=(8, 5))
plt.bar(valores, conteo, color='red', edgecolor='black')

plt.xlabel("Número del dado")
plt.ylabel("Frecuencia")
plt.title(f"Frecuencia de resultados en {1000} lanzamientos")
plt.xticks(range(1, 7))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Ejercicio 4. Gráfica de funciones trigonométricas.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

seno = np.sin(x)
coseno = np.cos(x)
tangente = np.tan(x)

plt.plot(x, seno, label="Seno", color="r")
plt.plot(x, coseno, label="Coseno", color="g")
plt.plot(x, tangente, label="Tangente", color="b")


plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Seno, Coseno y Tangente")
plt.legend()
plt.grid()
plt.show()
```

Ejercicio 5. Raices cuadradas.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1= np.sqrt(x)
y2 = x

plt.plot(x, y1, label="y = √x")
plt.plot(x, y2, label="y = x", linestyle="dashed")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de y = √x y de y = x")
plt.legend()
plt.grid()
plt.show()
```

Ejercicio 6. Pendiente de una recta.

```python
import matplotlib.pyplot as plt

X1 = float(input("Porfavor ingresa X1: "))
Y1 = float(input("Porfavor ingresa Y1: "))
X2 = float(input("Porfavor ingresa X2: "))
Y2 = float(input("Porfavor ingresa Y2: "))

pendiente = (Y2 - Y1) / (X2 - X1)
print(f"La pendiente es: {pendiente}")

valx = [X1, X2]
valy = [Y1, Y2]

plt.plot(valx, valy, marker="o", label="Recta")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfico Recta")
plt.legend()
plt.grid()
plt.show()
```

Ejercicio 7. Movimiento Parabólico.

```python
import numpy as np
import matplotlib.pyplot as plt


graved = 9.81

velinicial = float(input("Porfavor ingresa la velocidad inicial del proyectil: "))
angulo = float(input("Porfavor ingresa el ángulo de lanzamiento: "))

angulo = np.radians(angulo)


tiemtotal = (2 * velinicial * np.sin(angulo)) / graved
tiemp = np.linspace(0, tiemtotal, num=100)

x = velinicial * np.cos(angulo) * tiemp
y = velinicial * np.sin(angulo) * tiemp - 0.5 * graved * tiemp**2

plt.plot(x, y)
plt.xlabel("Distancia: ")
plt.ylabel("Altura: ")
plt.title("Trayectoria Proyectil")
plt.grid()
plt.show()
```
